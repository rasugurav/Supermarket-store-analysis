```python
data.drop(['Store ID '],axis=1,inplace=True)
```

```python
data.rename(columns={'Store_Area':'Area','Items_Available':'Items','Daily_Customer_Count':'Customer','Store_Sales':'Sales'},inplace=True)
```

```python
# there is a 100% coorelation between Area and Items so we can choose between one of them as they shows multicollinearrity .
#We will try both ways after considering both Area and Items and Considering only Items
```

```python
X=data.drop(['Sales'],axis=1)
Y=data['Sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr_predict=lr.predict(X_test)
```

```python
lr_predict
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
lr_MSE=mean_squared_error(lr_predict,y_test)
lr_MAE=mean_absolute_error(lr_predict,y_test)
lr_r2=r2_score(lr_predict,y_test)
```

```python
print("Mean squared error is:",lr_MSE)
print("Mean Absolute error is :",lr_MAE)
print("R2_score is:",lr_r2)
```

```python
# As model's error is high we can try more algoriths
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model=RandomForestRegressor()
```

```python
rf_model.fit(X_train,y_train)
```

```python
rf_predict=rf_model.predict(X_test)
```

```python
RF_MSE=mean_squared_error(rf_predict,y_test)
RF_MAE=mean_absolute_error(rf_predict,y_test)
RF_r2=r2_score(rf_predict,y_test)
```

```python
print('Mean Squared Error is:',RF_MSE)
print('Mean Absolute Error is:',RF_MAE)
print('r2 score is:',RF_r2)
```

```python
df=pd.read_csv('../input/stores-area-and-sales-data/Stores.csv')
```

```python
df.head()
```

```python
df.drop(['Store_Area','Store ID '],axis=1,inplace=True)
```

```python
df.head()
```

```python
X=df.drop(['Store_Sales'],axis=1)
Y=df['Store_Sales']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
predict_lr=lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
MSE_lr=mean_squared_error(predict_lr,y_test)
MAE_lr=mean_absolute_error(predict_lr,y_test)
r2_lr=r2_score(predict_lr,y_test)
print("Mean Squared error:",MSE_lr)
print("Mean Absolute error:",MAE_lr)
print("r2_score:",r2_lr)
```

```python
# error is similar to previous one
```
